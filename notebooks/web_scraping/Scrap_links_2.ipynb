```python
import pandas as pd
import numpy as np
import re
import time
import requests as rq
import bs4 as bs4
```

```python
df = pd.read_json("parsed_links.json", lines = True)
print('df shape:',df.shape)
df.head()
```

```python
# Select unique link
link_list = df['link'].unique()
len(link_list)
```

```python
# User- agent for get response
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}

for link in link_list:
    
    print(link)
    response = rq.get(link, headers = headers)

    link_name = re.search('carros-vans-e-utilitarios/(.*)',link).group(1)
    
    with open("./raw_data/links/link_{}.html".format(link_name), "w+") as output:
        output.write(response.text)
    time.sleep(2)
    
```
